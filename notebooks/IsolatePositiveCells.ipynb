```python
# Imports
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
def create_folder(folder_path):
    '''Creates a folder if does not exists.'''
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
```

```python
def crop_image(img, crop_size, left, top):
    '''crops the image in the defined size'''
    right = left +  crop_size
    bottom =  top + crop_size
    cropped = img.crop((left, top, right, bottom))
    return cropped
```

```python
def visualize(figsize=(15, 15),**images):
    """Plots images in one row."""
    n = len(images)
    plt.figure(figsize=figsize)
    for i, (name, image) in enumerate(images.items()):
        plt.subplot(1, n, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.title(' '.join(name.split('_')).title())
        plt.imshow(image)
    plt.show()
```

```python
create_folder('../IsolatedCellsDataset/original/all/1')
```

```python
image_01 = Image.open('../CellsDataset/segmented/1.png')
image_02 = Image.open('../CellsDataset/segmented/2.png')
image_03 = Image.open('../CellsDataset/segmented/3.png')
image_04 = Image.open('../CellsDataset/segmented/4.png')
image_05 = Image.open('../CellsDataset/segmented/5.png')
visualize(
    complete_image_01=image_01,
    complete_image_02=image_02,
    complete_image_03=image_03,
    complete_image_04=image_04,
    complete_image_05=image_05   
)
```
